```python
# Passo 4: Análise inicial (E.g.: Quantos clientes cancelaram)

# Quantidade - Contar quantos são 0 e quantos são 1 na coluna CANCELOU
display(table["cancelou"].value_counts())

# Porcentagem
display(table["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```
